```python
data = pd.read_csv('/kaggle/input/students-scores/student_scores.csv')

```

```python
x=data['Hours'].values
y=data['Scores'].values
```

```python
plt.scatter(x,y)
plt.xlabel('hours')
plt.ylabel('scores')
plt.title('data discription')
plt.show()
```

```python
l_rate=0.001
iteration=100
theta1=0
theta0=0
n=x.shape[0]
losses=[]
for i in range(iteration):
    h_x=theta0+theta1*x
    mse=(1/n)*np.sum((h_x-y)**2)
    losses.append(mse)
    d_theta0=(2/n)*np.sum(h_x-y)
    d_theta1=(2/n)*np.sum(x*(h_x-y))
    theta1=theta1- l_rate*d_theta1
    theta0=theta0 - l_rate*d_theta0
        
    
```

```python
print('theta0',theta0)
print('theta1',theta1)
print('MSE',mse)
```

```python
newx=9
predict=theta0+theta1*newx
print(predict)
```

```python
plt.plot(losses)
plt.xlabel('iteration')
plt.ylabel('losses')
```

```python
x_line=np.linspace(0,10,100)
y_line=theta0+theta1*x_line
plt.plot(x_line,y_line,c='r')
plt.scatter(x,y,s=10)
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```
