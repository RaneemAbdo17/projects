```python
my_string=" python is a fancy programming language"
m1=re.findall(r"\s\w\w\w\w\w",my_string) #white space followed by 5 characters
print(m1)
m2=re.findall(r"\s\w*",my_string)#white space followed by zero or more characters
print(m2)
```

```python
m3=re.findall(r"[pP]\w\w",my_string)
print(m3)
```

```python
m4=re.findall(r".*\s",my_string)
print(m4)
```

```python
bee_data=pd.read_csv("https://raw.githubusercontent.com/CSB-book/CSB/master/regex/data/bee_list.txt",sep="\t")
```

```python
bee_data.info()
```

```python
bee_data.head()
```

```python
authors=bee_data['Taxon Author']
len(authors)
```

```python
my_reg=re.compile(r"\(?([\w\s,\.\-\&]*),\s(\d{4})\)?")  #\(? open parenthesis  \)? close  
```

```python
def extract_list_au_year(au):
    tmp=re.match(my_reg,au)
    author_list=tmp.group(1)
    year=tmp.group(2)
    author_list=re.split(", | \&",author_list)
    return[author_list,year]
```

```python
x=authors[44]
extract_list_au_year(x)
```
