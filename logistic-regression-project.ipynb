```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
diabetes=pd.read_csv('https://raw.githubusercontent.com/DrSaadLa/PythonTuts/main/GLM_Models/Data/diabetes.csv')
```

```python
diabetes.info()
```

```python
diabetes.describe().T
```

```python
diabetes.isnull().sum()
```

**data preprocessing**

```python
diabetes_c=diabetes.copy(deep=True)
inv_var=['Glucose','BloodPressure','SkinThickness','Insulin','BMI']
diabetes_c[inv_var]=diabetes_c[inv_var].replace(0,np.NaN)
print(diabetes_c.isnull().sum())
```

```python
sns.heatmap(diabetes_c.isnull(),cbar=False)
```

```python
_=diabetes.hist(figsize=(20,20))
```

**missing data imputation**

```python
diabetes_c['Glucose'].fillna(diabetes_c['Glucose'].mean(),inplace=True)
diabetes_c['BloodPressure'].fillna(diabetes_c['BloodPressure'].mean(),inplace=True)
diabetes_c['SkinThickness'].fillna(diabetes_c['SkinThickness'].median(),inplace=True)
diabetes_c['Insulin'].fillna(diabetes_c['Insulin'].median(),inplace=True)
diabetes_c['BMI'].fillna(diabetes_c['BMI'].median(),inplace=True)
diabetes_c.describe().T
```

```python
_=diabetes_c.hist(figsize=(20,20))
```

```python
sns.heatmap(diabetes_c.corr(),annot=True)
```

**standardization**

```python
from sklearn.preprocessing import StandardScaler
std_sc=StandardScaler()
```

```python
x=diabetes_c.drop(['Outcome'],axis=1)
x=std_sc.fit_transform(x)
y=diabetes_c.Outcome
x_df=pd.DataFrame(x,columns=diabetes_c.columns[:-1])
x_df.describe().T
```

```python
X=x_df.values
y=diabetes_c['Outcome'].values
print(X.shape)
print(y.shape)
```

**split data into train test**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.75,test_size=0.25,random_state=10123)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,classification_report
logreg=LogisticRegression()
logreg.fit(X_train,y_train)
y_pred=logreg.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

**KNN classifier**

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=6)
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
