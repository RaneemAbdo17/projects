```python
import numpy as np
```

**manual calculation**

```python
input_layer=np.array([9,7])
weights={"node0":np.array([3,7]),"node1":np.array([-5,-2]),"output":np.array([3,9])}
```

```python
def relu(input):
    output=max(0,input)
    return output
```

```python
node0_value=(input_layer*weights['node0']).sum()
node0_actual=relu(node0_value)
node1_value=(input_layer*weights['node1']).sum()
node1_actual=relu(node1_value)
hidden_layer=np.array([node0_actual,node1_actual])
output=(hidden_layer*weights['output']).sum()
output_actual=relu(output)
```
