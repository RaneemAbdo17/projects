```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_layer=np.array([0,3])
weights={"node0":np.array([2,1]),"node1":np.array([1,2]),"output":np.array([1,1])}
target_actual=3
```

```python
def relu(input):
    output=max(0,input)
    return output
```

```python
def one_hidden_layer_network(input_layer,weight):
    node0=(input_layer*weight['node0']).sum()
    node0_value=relu(node0)
    node1=(input_layer*weight['node1']).sum()
    node1_value=relu(node1)
    hidden_layer=np.array([node0_value,node1_value])
    output=(hidden_layer*weight['output']).sum()
    output_value=relu(output)
    return output_value
```

```python
model_output_0=one_hidden_layer_network(input_layer,weights)
```

```python
error0=model_output_0-target_actual
print(error0)
```

```python
weights_1={"node0":np.array([2,1]),"node1":np.array([1,2]),"output":np.array([-1,1])}
model_output_1=one_hidden_layer_network(input_layer,weights_1)
error1=model_output_1-target_actual
print(error0)
print(error1)
```

```python
input_data=[np.array([0,3]),np.array([1,2]),np.array([-1,-2]),np.array([4,0])]
target_actual=np.array([1,3,5,7])
weights_0={"node0":np.array([2,1]),"node1":np.array([1,2]),"output":np.array([1,1])}
weights_1={"node0":np.array([2,1]),"node1":np.array([1,1.5]),"output":np.array([1,1.5])}
```

```python
from sklearn.metrics import mean_squared_error
```

```python
model_output_0=[]
model_output_1=[]
for data in input_data:
    model_output_0.append(one_hidden_layer_network(data,weights_0))
    model_output_1.append(one_hidden_layer_network(data,weights_1))
mse_0=mean_squared_error(target_actual,model_output_0)
mse_1=mean_squared_error(target_actual,model_output_1)
    
print("mean squared error with weight 0",mse_0)
print("mean squared error with weight 1",mse_1)
```

**gradient descent**

```python
weights=np.array([0,2,1])
input_data=np.array([1,2,3])
target=0

#calculate prediction
preds=(weights*input_data).sum()

#calculate error
error=preds-target

#calculate slope = gradient descent
slope=2*input_data*error

print(slope)
```

```python
#improve the model with learning rate
learning_rate=0.01
preds=(weights*input_data).sum()
error=preds-target
slope=2*input_data*error
weights_update=weights-(slope*learning_rate)
preds_update=(weights_update*input_data).sum()
error_update=preds_update-target
slope=2*input_data*error_update
print("original error =",error)
print("updated error =",error_update)
print("gradient descent =",slope)
```

**multiple learning rate**

```python
def get_error(input_data,target,weights):
    preds=(weights*input_data).sum()
    error=preds-target
    return(error)

def get_slope(input_data,target,weights):
    error=get_error(input_data,target,weights)
    slope=2*error*input_data
    return(slope)

def get_mse(input_data,target,weights):
    errors=get_error(input_data,target,weights)
    mse=np.mean(errors**2)
    return(mse)

n_updates=20
mse_hist=[]
for i in range(n_updates):
    slope=get_slope(input_data,target,weights)
    weights=weights-0.01*slope
    mse=get_mse(input_data,target,weights)
    mse_hist.append(mse)
    
plt.plot(mse_hist)
plt.xlabel("iterations")
plt.ylabel("mean squared error")
plt.show()
```
