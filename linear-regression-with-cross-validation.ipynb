```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
```

```python
 data_url = "http://lib.stat.cmu.edu/datasets/boston"
 raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
 data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
 target = raw_df.values[1::2, 2]
```

```python
df=pd.DataFrame(data,columns=['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT'])
```

```python
df['target']=target
```

```python
from sklearn.model_selection import KFold
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,make_scorer
lm=LinearRegression()
```

 **Manual Cross Validation**

```python
kf=KFold(n_splits=5,shuffle=True,random_state=10123)
splits=kf.split(data)
#access training and validation indicies
for train_index,val_index in splits:
    #split data into train and validation sets
    x_train,y_train=data[train_index],target[train_index]
    x_val,y_val=data[val_index],target[val_index]
    #fit linear regression
    lm.fit(x_train,y_train)
    #make prediction
    preds=lm.predict(x_val)
    print("mean squared error ",mean_squared_error(y_val,preds))
    print("mean absolute error ",mean_absolute_error(y_val,preds))
    print()
```

```python
from sklearn.model_selection import cross_val_score
cv_results=cross_val_score(lm,data,target,cv=5)
print(cv_results)
```

```python
avg=np.mean(cv_results)
print("the average r2 ",avg)
```

```python
mae=make_scorer(mean_absolute_error)
cv_res=cross_val_score(lm,data,target,cv=5,scoring=mae)
print(cv_res)
```

```python
average=np.mean(cv_res)
print(average)
```

```python
mse=make_scorer(mean_squared_error)
res=cross_val_score(lm,data,target,cv=5,scoring=mse)
print(res)
```

```python
avgg=np.mean(res)
print(avgg)
```
