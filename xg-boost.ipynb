```python
from xgboost import XGBClassifier
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
from sklearn.model_selection import train_test_split,cross_val_score,StratifiedKFold,RandomizedSearchCV,GridSearchCV
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score,roc_auc_score,classification_report,confusion_matrix
```

```python
hd=pd.read_csv('https://raw.githubusercontent.com/DrSaadLa/PythonTuts/main/ML_and_DL%20Workshop/heart_disease.csv')
```

```python
hd.info()
```

```python
x=hd.iloc[:,:-1]
y=hd.iloc[:,-1]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=101)
```

```python
xgb_clf=XGBClassifier(booster='gbtree',objective='binary:logistic',random_state=101,eval_metric='logloss')
```

```python
xgb_clf.fit(x_train,y_train)
pred=xgb_clf.predict(x_test)
score=accuracy_score(y_test,pred)
print("the accuracy score ",score)
```

```python
cv_scores=cross_val_score(xgb_clf,x,y,cv=5)
accuracy=np.round(cv_scores,2)
mean=cv_scores.mean()*100
print("accuracy ",accuracy)
print("mean of accuracy ",mean)
```

```python
kfold=StratifiedKFold(n_splits=5,shuffle=True,random_state=101)
scores=cross_val_score(xgb_clf,x,y,cv=kfold)
accuracy=np.round(scores,2)
mean=cv_scores.mean()
print("accuracy ",accuracy)
print("mean of accuracy ",mean)
```
