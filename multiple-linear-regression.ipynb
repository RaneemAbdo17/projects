```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
```

```python
 data_url = "http://lib.stat.cmu.edu/datasets/boston"
 raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
 data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
 target = raw_df.values[1::2, 2]
```

```python
df=pd.DataFrame(data,columns=['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT'])
```

```python
df['target']=target
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
```

```python
lm.fit(data,target)
```

```python
print("the model intercept value is {:.3f}".format(lm.intercept_))
```

```python
for(var,coef) in list(zip(df[:0],lm.coef_)):
    print("the coefficient of",var," is " ,coef)
```

```python
result=pd.DataFrame(list(zip(df[:0],lm.coef_)),columns=["variables","coefficient"])
result
```

```python
pred=lm.predict(data)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
print("mean squared error",mean_squared_error(target,pred))

print("mean absolute error",mean_absolute_error(target,pred))

print("r2 error",r2_score(target,pred))

print("root mean squared error",np.sqrt(mean_squared_error(target,pred)))
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(data,target,test_size=0.2,random_state=123)
```

```python
lm.fit(X_train,y_train)
```

```python
pred_train=lm.predict(X_train)
pred_test=lm.predict(X_test)
```

```python
print("mean squared error train set",mean_squared_error(y_train,pred_train))
```

```python
print("mean squared error test set",mean_squared_error(y_test,pred_test))
```

```python
print("root mean squared error in train set",np.sqrt(mean_squared_error(y_train,pred_train)))
```

```python
print("root mean squared error in test set",np.sqrt(mean_squared_error(y_test,pred_test)))
```

```python
print("r2 score in train",r2_score(y_train,pred_train))
```

```python
print("r2 score in test",r2_score(y_test,pred_test))
```
