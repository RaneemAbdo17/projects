```python
x = data[['enginesize', 'carheight']].values
y = data['price'].values

print(x.shape)
print(y.shape)
```

```python
class LinearRegression:
    def __init__(self,l_rate=0.001,iteration=1000):
        self.l_rate=l_rate
        self.iteration=iteration
        
    def fitG(self,x,y):
        self.cost=[]
        self.theta = np.zeros((1 + x.shape[1]))
        n=x.shape[0]
        for i in range(self.iteration):
            y_pred=self.theta[0]+np.dot(x,self.theta[1:])
            mse=(1/n)*np.sum((y_pred-y)**2)
            self.cost.append(mse)
            d_theta0=(2/n)*np.sum(y_pred-y)
            d_theta1=(2/n)*np.dot(x.T,(y_pred-y))
            
            self.theta[0]=self.theta[0]-self.l_rate*d_theta0
            self.theta[1:]=self.theta[1:]-self.l_rate*d_theta1
            return self
    def predictG (self,x):
        return self.theta[0]+np.dot(x,self.theta[1:])
    
    def fitN(self,x,y):
        z = np.ones((x.shape[0],1))
        x=np.append(z,x,axis=1)
        self.thetas=np.linalg.pinv(x.T.dot(x)).dot(x.T).dot(y)
        return self
    def predictN(self,x):
        z = np.ones((x.shape[0],1))
        x=np.append(z,x,axis=1)
        return np.dot(x,self.thetas)
        
            
```

```python
def scale(x):
    x_scaled=x-np.mean(x,axis=0)
    x_scaled=x_scaled/np.std(x_scaled,axis=0)
    return x_scaled
```

```python
x_sd=scale(x)
lr=LinearRegression()
lr.fitG(x_sd,y)
lr.fitN(x,y)
```

```python
print('theta0',lr.theta[0])
print('theta1',lr.theta[1])
print('theta2',lr.theta[2])
```

```python
print('theta0',lr.thetas[0])
print('theta1',lr.thetas[1])
print('theta2',lr.thetas[2])
```

```python
ypred=lr.predictG(x_sd)
zpred=lr.predictN(x)
```

```python
errors = np.sum((ypred - y)**2)
sst = np.sum((y - np.mean(y))**2)
r2_GD = 1 - (errors/sst)

adusted_r2_GD= 1-((1-r2_GD) * (x.shape[0] -1)/(x.shape[0] - x.shape[1] - 1))
print (adusted_r2_GD)
```

```python
errors = np.sum((zpred - y)**2)
sst = np.sum((y - np.mean(y))**2)
r2_NQ = 1 - (errors/sst)

adusted_r2_NQ= 1-((1-r2_NQ) * (x.shape[0] -1)/(x.shape[0] - x.shape[1] - 1))
print (adusted_r2_NQ)
```
