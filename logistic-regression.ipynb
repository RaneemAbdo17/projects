```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url="https://raw.githubusercontent.com/DrSaadLa/PythonTuts/main/GLM_Models/Data/churn.csv"
```

```python
churn=pd.read_csv(url)
```

```python
churn.head()
```

```python
churn['has_churned'].value_counts()
```

```python
sns.displot(data=churn,x='time_since_first_purchase',col='has_churned')
plt.show()
```

```python
sns.regplot(x='time_since_last_purchase',y='has_churned',data=churn,ci=None)
plt.show()
```

```python
import statsmodels.api as sm 
from statsmodels.formula.api import glm
```

```python
first_mdl= glm("has_churned ~ time_since_first_purchase",data=churn,family = sm.families.Binomial()).fit()
```

```python
last_mdl = glm("has_churned ~ time_since_last_purchase",data=churn,family=sm.families.Binomial()).fit()
```

**prediction with logistic regression**

```python
new_explanatory_data=pd.DataFrame({"time_since_last_purchase":np.arange(-1,6.5,0.25)})
predicted_data=new_explanatory_data.assign(has_churned=last_mdl.predict(new_explanatory_data))
predicted_data.head()
```

```python
sns.regplot(x="time_since_last_purchase",y="has_churned",data=churn,logistic=True,ci=None)
sns.scatterplot(x="time_since_last_purchase",y="has_churned",data=predicted_data)
```

```python
predicted_data["most_likely_outcome"]=np.round(predicted_data['has_churned'])
predicted_data.head()
```

```python
predicted_data["odds_ratio"]=(predicted_data['has_churned']/(1-predicted_data['has_churned']))
predicted_data.head()
```

```python
sns.lineplot(x='time_since_last_purchase',y='odds_ratio',data=predicted_data)
plt.axhline(y=1,color='red')
plt.axvline(x=0,color='red')
```

```python
predicted_data['log_odds_ratio']=np.log(predicted_data['odds_ratio'])
predicted_data.head()
```

```python
sns.lineplot(x='time_since_last_purchase',y='log_odds_ratio',data=predicted_data)
plt.axhline(y=0,color='red')
plt.axvline(x=0,color='red')
```

```python
new_data=pd.DataFrame({"time_since_first_purchase":np.arange(-1,6.5,0.25)})
pred_data=new_data.assign(has_churned=first_mdl.predict(new_data))
pred_data.head()
```

```python
sns.regplot(x="time_since_first_purchase",y="has_churned",data=churn,logistic=True,ci=None)
sns.scatterplot(x="time_since_first_purchase",y="has_churned",data=pred_data)
```

```python
pred_data["most_likely_outcome"]=np.round(pred_data['has_churned'])
pred_data.head()
```

```python
pred_data["odds_ratio"]=(pred_data['has_churned']/(1-pred_data['has_churned']))
pred_data.head()
```

```python
pred_data['log_odds_ratio']=np.log(pred_data['odds_ratio'])
pred_data.head()
```

**make confusion matrix**

```python
actual_response=churn['has_churned']
predicted_response=np.round(first_mdl.predict())
outcomes=pd.DataFrame({"actual_response":actual_response,"predicted_response":predicted_response})
print(outcomes.value_counts(sort=False))
```

```python
from statsmodels.graphics.mosaicplot import mosaic
conf_matrix=np.reshape(outcomes.value_counts(sort=False).values,(2,2))
mosaic(conf_matrix)
plt.show()
```

```python
TN=conf_matrix[0,0]
TP=conf_matrix[1,1]
FN=conf_matrix[1,0]
FP=conf_matrix[0,1]
```

```python
accuracy=(TN+TP)/(TN+TP+FN+FP)
sesitivity=TP/(FN+TP)
specificity=TN/(TN+FP)
print("accuracy = ",accuracy)
print("sensitivity = ",sesitivity)
print("specificity = ",specificity)
```
